## Dimensionality Reduction with PCA

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
print y[np.argsort(y)]
```

```python
plt.imshow(X[np.argsort(y),], aspect=0.02, cmap=plt.cm.gray_r)
plt.show()
```

```python
pca = PCA(n_components=64).fit(X)
X_reduced =pca.transform(X)
```

```python
print X_reduced.shape
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.hlines(0.8, 0, 64, color='lightgrey')
plt.show()
```

```python
plt.imshow(X_reduced[np.argsort(y),], aspect=0.02, cmap=plt.cm.gray_r)
plt.show()
```

```python
plt.scatter(X_reduced[:,0], X_reduced[:,1], c=y, s=50, alpha=0.7)
plt.show()
```

```python




kmeans = KMeans(n_clusters=2)



def plot_kmeans(X, y):
    y_pred = kmeans.fit(X).predict(X)
    fig, ax = plt.subplots(figsize=(6,6))

    ax.scatter(X[:, 0], X[:, 1], c=y, s=70, alpha=0.5, edgecolor='none')
    
    print(abs(y_pred-y))
    print(kmeans.cluster_centers_)
    
#     ax.scatter(X[y!=y_pred, 0], X[y!=y_pred, 1], c='black' , s=70, 
#                alpha=1.0, edgecolor='none')

    
    ax.set_xlabel('X[:,0]')
    ax.set_ylabel('X[:,1]')
    fig.show()
```

```python
X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=1 )

y_pred = kmeans.fit(X).predict(X)
plot_kmeans(X,y)
```
